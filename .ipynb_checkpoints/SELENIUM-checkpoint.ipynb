```python
ls = driver.find_elements(By.CLASS_NAME, 'RfBGI')
len(ls)
```

```python
nombre = []
puntuacion = []

#descripcion = []
#precio = []
tipo_cocina = []
ubicacion = []



for e in ls:
    time.sleep(2)
    e.click()
    driver.switch_to.window(driver.window_handles[1])
    time.sleep(2)

    nombre.append(driver.find_element(By.CLASS_NAME, 'HjBfq').text)

    puntuacion.append(driver.find_element(By.CLASS_NAME, 'ZDEqb').text)

    tipo_cocina.append([e.text for e in driver.find_elements(By.CLASS_NAME, 'dlMOJ')])

    ubicacion.append([e.text for e in driver.find_elements(By.CLASS_NAME, 'AYHFM')][1])


    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    print(len(nombre))
```
