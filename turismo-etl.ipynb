```python
import requests as req
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import feedparser
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from pymongo import MongoClient
from selenium.webdriver.chrome.options import Options
import time
import tqdm
import warnings
warnings.filterwarnings('ignore')

PATH = ChromeDriverManager().install()
```

```python
client = MongoClient('mongodb://localhost:27017')
db = client['turismo']
collection = db['turismo-madrid']
```

```python
driver = webdriver.Chrome(PATH)
```

```python
driver.get('https://www.tripadvisor.es/Attractions-g187514-Activities-oa930-Madrid.html')
```

```python
wait = WebDriverWait(driver, 15)
wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="onetrust-accept-btn-handler"]'))).click()
```

```python
contador = 0

while 1:

    ls = driver.find_elements(By.CLASS_NAME, 'XfVdV.o.AIbhI') #ok
    
    for e in ls:
        
        time.sleep(2)
        
        try:
            try:
                e.click() #ok
            
            except:
                driver.execute_script('window.scrollBy(0, 200);')
                e.click()
        except:
            driver.execute_script('window.scrollBy(0, 150);')
            e.click()
            
        
        time.sleep(1)  
        driver.switch_to.window(driver.window_handles[1])
        driver.execute_script("return document.readyState == 'complete';")
        

        try:
            time.sleep(4)
            name = driver.find_element(By.CLASS_NAME, 'biGQs._P.fiohW.eIegw').text #ok

        except:
            driver.refresh()
            #driver.execute_script("return document.readyState == 'complete';")
            time.sleep(4)
            name = driver.find_element(By.CLASS_NAME, 'biGQs._P.fiohW.eIegw').text
            
        try:
            link = driver.find_element(By.CLASS_NAME, 'UikNM._G.B-._S._T.c.G_.P0.wSSLS.wnNQG.raEkE').get_attribute('href') #ok
        except:
            link = 'No hay link'
        
        try: #OK
            ubicacion = driver.find_element(By.XPATH, '//*[@id="tab-data-WebPresentation_PoiLocationSectionGroup"]/div/div/div[2]/div[1]/div/div/div/div[1]/button/span').text
        except:
            ubicacion = 'no disponible'
        
        time.sleep(2)
        try:
            categoria = driver.find_element(By.CLASS_NAME, 'fIrGe._T.bgMZj').text 

        except:
            categoria = driver.find_element(By.CLASS_NAME, '_T.FKffI').text
        
        driver.execute_script('window.scrollTo(0, 2700);')
        try: #OK
            puntuacion = driver.find_element(By.CLASS_NAME, 'biGQs._P.fiohW.hzzSG.uuBRH').text 
        except:
            puntuacion = 'Esta actividad todavía no tiene puntuación, ¡se el primero!'
        
       
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

        turismo = {'_id':name, 'nombre': name, 'link':link, 'ubicacion':ubicacion,
                   'puntuacion':puntuacion, 'categoria':categoria}
        print(name, link, ubicacion, categoria, puntuacion)
        
        try:
            collection.insert_one(turismo)
            contador += 1
            print(contador)
            
        except:
            continue
    
    try:
        driver.execute_script('window.scrollBy(0, 400);')
        driver.find_element(By.XPATH, '//html/body/div[1]/main/div[1]/div/div[3]/div/div[2]/div[2]/div[2]/div/div[3]/div/div[2]/div/div/section[41]/div/div[1]/div/div[1]/div[2]/div/a').click()
        driver.execute_script("return document.readyState == 'complete';")
        time.sleep(1)

    except:
        break

```
